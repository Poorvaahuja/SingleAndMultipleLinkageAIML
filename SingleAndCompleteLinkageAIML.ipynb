```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.datasets import make_blobs
```

```python
np.random.seed(42)
X, _ = make_blobs(n_samples=100, centers=3, cluster_std=1.5, random_state=42)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c='blue', marker='o', s=50)
plt.title("Generated Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# Perform hierarchical clustering using single linkage
single_linkage = linkage(X, method='single')
complete_linkage = linkage(X, method='complete')
```

```python
# Plot dendrogram for single linkage
plt.figure(figsize=(12, 6))
plt.title("Hierarchical Clustering Dendrogram (Single Linkage)")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
dendrogram(single_linkage, leaf_rotation=90., leaf_font_size=8.)
plt.show()
```

```python
# Plot dendrogram for complete linkage
plt.figure(figsize=(12, 6))
plt.title("Hierarchical Clustering Dendrogram (Complete Linkage)")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
dendrogram(complete_linkage, leaf_rotation=90., leaf_font_size=8.)
plt.show()
```
